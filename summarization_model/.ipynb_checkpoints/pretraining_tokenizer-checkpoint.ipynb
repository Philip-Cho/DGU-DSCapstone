# 데이터 전처리
1) corpus dataset에서 필요한 부분 추출  
2) 데이터별로 리스트에 저장, txt file로 저장  
3) 문장 단위로 리스트에 다시 저장  
4) 전처리(lower case, 공백 제거, 특수문자 제거 등)

```python
import jsonlines
import json
import re
import string
import nltk
from nltk.data import load
```

```python
with open('./data/open_research/open_research_sample.json') as file:
    data = json.load(file)
```

```python
[a['fieldsOfStudy'] for a in data]
```

```python
data[0]
```

```python
# 'paperAbstract'에서 공백 제외한 텍스트 -> 리스트
data_txt = [a['paperAbstract'] for a in data if a['paperAbstract']!='']
data_txt
```

```python
# 여러 문장으로 이뤄진 글 -> 문장별로 나눠서 txt 파일로 저장 1
with open('./data/open_research/sample_parag.txt', 'w', encoding='utf-8') as file:
    for line in data_txt:
        sents = line.split('. ')
        for sent in sents:
            if sent[0].isupper():
                file.write('\n'+sent)
            else:
                file.write('. '+sent)
```

```python
# 저장한 txt 파일 읽어서 빈줄, 줄바꿈 제거
with open('data/open_research/sample_parag.txt', 'r', encoding='utf-8') as file:
    data = file.readlines()
data = [line.replace('\n', '') for line in data]
data = [line for line in data if line!='']
data
```

```python
# lower case로 바꿔줌
def do_lower(lst):
    after_lower = [line.lower() for line in lst]
    return after_lower

data = do_lower(data)
data
```

```python
def remove_special(lst):
    t = list(map(lambda x: re.sub(r"(\$|€|¥|₹|£)", "", x), lst))
    t = list(map(lambda x: re.sub(r"\d+", "", x), t))
    t = list(map(lambda x: re.sub(r"\%|:|;|-|_|《|》|<|>|—|⋯|—", "", x), t))
    t = list(map(lambda x: x.replace('"|\'', ''), t))
    t = list(map(lambda x: x.translate(str.maketrans('','', string.punctuation)), t))
    t = list(map(lambda x: re.sub(' +', ' ', x), t))
#     t = list(map(lambda x: re.sub(' ', '', x), t))
    return t

data = remove_special(data)
data
```

```python
def remove_whitespace(lst):
    t = [line for line in lst if line!=' ']
    t = [line for line in t if line!='\n']
    return t

data = remove_whitespace(data)
data
```

```python
with open('./data/open_research/sample_sents.txt', 'w', encoding='utf-8') as file:
    for line in data:
        file.write(line+'\n')
```

# tokenizer 학습, 저장

```python
from pathlib import Path
from tokenizers import ByteLevelBPETokenizer
from transformers import BartTokenizer
from tokenizers.processors import BertProcessing

tokenizer = ByteLevelBPETokenizer()

tokenizer.train(files='data/separate_sents(0~999).txt',
               vocab_size=32000,
               min_frequency=2,
               special_tokens=['<pad>', '<unk>', '<s>', '</s>', '<mask>'],
               )

tokenizer.save_model('output')
tokenizer = BartTokenizer(vocab_file='output/vocab.json', merges_file='output/merges.txt', do_lower_case=True)
tokenizer.save_pretrained('output/tokenizer_ord')


trained_tokenizer = BartTokenizer.from_pretrained('output/tokenizer_ord')
enc = trained_tokenizer.encode('This is a newly-pretrained tokenizer!')
print(trained_tokenizer.decode(enc))
```
